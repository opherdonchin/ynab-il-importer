# Inspect Card Account Mapping

This notebook inspects normalized card CSVs to verify `source_account` and `account_name`.

```python
from pathlib import Path
import pandas as pd

paths = sorted(Path("data/derived").glob("card_normalized*.csv"))
paths
```

```python
for path in paths:
    df = pd.read_csv(path)
    cols = [c for c in ["source_account", "account_name"] if c in df.columns]
    print("\n", path.name)
    if not cols:
        print("missing account columns")
        continue
    pairs = df[cols].dropna().drop_duplicates()
    print(pairs.to_string(index=False))
```
